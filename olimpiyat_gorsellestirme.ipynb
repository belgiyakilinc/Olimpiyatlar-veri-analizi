## Renk Paletleri

1. Nitel (Qualitative) Paletler
Bu paletler, aralarında doğal bir sıra veya ilişki olmayan kategorik verileri (örneğin, meyve türleri, ülkeler, cinsiyet) görselleştirmek için uygundur. Her kategori farklı, kolayca ayırt edilebilir bir renkle temsil edilir.

Varsayılan ve Varyasyonları (tab10):

'deep' (Varsayılan)

'muted'

'pastel'

'bright'

'dark'

'colorblind' (Renkli görme kusurları olanlar için daha uygun)

Color Brewer Nitel Paletleri: 'Set1', 'Set2', 'Set3', 'Paired', 'Accent', 'Dark2', 'Pastel1', 'Pastel2'.

Döngüsel Paletler: 'husl', 'hls'.

2. Sıralı (Sequential) Paletler
Bu paletler, bir aralıkta sıralanmış veya sürekli sayısal verileri (örneğin, sıcaklık, gelir düzeyi) görselleştirmek için idealdir. Renkler genellikle açık bir tondan koyu bir tona veya tersine doğru ilerler.

Algısal Olarak Tekdüze (Perceptually Uniform):

'rocket', 'mako' (Yoğunluk haritaları için iyi)

'flare', 'crest'

Matplotlib'den gelenler: 'viridis', 'plasma', 'inferno', 'magma'

Color Brewer Sıralı Paletleri: 'Blues', 'Greens', 'Reds', 'Oranges', 'Purples' vb. (Tek renk tonlu) veya 'BuGn', 'YlOrRd' vb. (Çok renk tonlu).

3. Iraksak (Diverging) Paletler
Bu paletler, sıfır veya bir eşik değeri gibi belirgin bir merkez noktası olan sayısal verileri (örneğin, pozitif ve negatif değerler, ortalamadan sapma) göstermek için kullanılır. Merkezde nötr bir renk (genellikle beyaz veya gri) bulunur ve bu merkezden iki farklı renge doğru açık tonlardan koyu tonlara doğru ayrılır.

Color Brewer Iraksak Paletleri: 'RdBu', 'PuOr', 'BrBG', 'PRGn', 'PiYG' vb.

```python
ozel_palet = ['yellow', "brown" , 'gray']

sns.scatterplot(x="boy", y = "kilo", hue="madalya", data=veri, palette=ozel_palet)
plt.title("Madalyaya Göre Boy ve Ağırlık İlişkisinin Saçılım Grafiği")
plt.show()
```

## Histogram

```python
sns.displot(veri, x=("kilo"))
plt.ylabel("Frekans")
plt.xlabel("Ağırlık")
plt.title("Ağırlık Dağılımı")
plt.show()
```

```python
sns.displot(veri, x=("boy"), hue="cinsiyet")
plt.ylabel("Frekans")
plt.xlabel("Boy")
plt.title("Boy Dağılımı")
plt.show()
```

```python
sns.displot(veri, x=("boy"), col="cinsiyet",multiple="dodge")
plt.show()
```

```python
sns.displot(veri, x=("kilo"), y=("boy"), hue= "cinsiyet", kind="kde") #Kernel Density Estimation - Çekirdek Yoğunluk Tahmini(olasılık dağılım fonksiyonunu tahmin etmek)
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Ağırlık ve Boy Dağılımı")
plt.show()
```

```python
sns.kdeplot(data=veri, x=("kilo"), y=("boy"), hue= "cinsiyet", fill=True) #Kernel Density Estimation - Çekirdek Yoğunluk Tahmini(olasılık dağılım fonksiyonunu tahmin etmek)
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Ağırlık ve Boy Dağılımı")
plt.show()
```

```python
sns.barplot(x="madalya", y="kilo", data=veri,hue="cinsiyet")
plt.title("Madalyaya Göre Ağırlık Dağılımı")
plt.show()
```

```python
sns.catplot(x="madalya", y="yas",hue="cinsiyet", data=veri, col="sezon",kind="bar", height=4, aspect=0.7)
plt.show()
```

## Isı haritası

```python
columns_to_drop = ['isim', 'cinsiyet', 'takim', 'uok', 'sezon', 'sehir', 'spor', 'etkinlik', 'madalya']
veri_numeric = veri.drop(columns=columns_to_drop)
sns.heatmap(veri_numeric.corr(), annot=True, linewidths=0.5, fmt=".2f")
plt.show()
```

## Ortaklık Grafiği

```python
sns.jointplot(data=veri, x="boy", y="kilo", hue="sezon", kind="kde"),
plt.show()
```

```python
g = sns.JointGrid(data=veri, x="kilo", y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
sns.pairplot(veri)
plt.show()
```

```python
g = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
plt.show()
```

```python
sns.countplot(data=veri, x="sehir")
plt.xticks(rotation=90)
plt.show()
```
